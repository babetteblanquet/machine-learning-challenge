```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
#Convert string values into integer
df = pd.get_dummies(df)
df.head()
```

```python
#Drop unnecessary column resulting from get_dummy
df = df.drop("koi_disposition_CANDIDATE", axis=1)
df.head()
```

# Select your features (columns)

```python
#Drop Koi_disposition as it is our y values and also to be able to perform a Variance treshold analysis:
#data = df.drop("koi_disposition", axis=1)
```

```python
#Identify variables with low variance (<0.1) to retain only meaningful variables:
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0.1)
var_thres.fit(df)
```

```python
var_thres.get_support()
```

```python
#Return the columns with a variance higher than 0.1
df.columns[var_thres.get_support()]
```

```python
#Return the number of columns that have a variance higher than 0.1
len(df.columns[var_thres.get_support()])
```

```python
# Set features. This will also be used as your x values.
selected_features = df[df.columns[var_thres.get_support()]]
selected_features.head()
```

```python
#Assigning the selected features to a variable named data:
data = selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign X (data) and y (target)
y = data[["koi_disposition_CONFIRMED","koi_disposition_FALSE POSITIVE"]].values.reshape(-1, 2)
X = data.drop(labels=["koi_disposition_CONFIRMED","koi_disposition_FALSE POSITIVE"], axis=1)
print(X.shape, y.shape)
```

```python
#Create a Train Test Split:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
# Create a MinMaxScaler model and fit it to the training data
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
y_scaler = MinMaxScaler().fit(y_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
#Create a RandomForest model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train_scaled)
rf.score(X_test_scaled, y_test_scaled)
```

```python
#Feature importance analysis
feature_names = data.columns
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
#Dropping the variables that have less than 2% influence
data2 = data.drop(labels=['koi_impact', 'koi_duration', 'koi_teq', 'koi_srad_err1', 'koi_insol', 'koi_depth_err2', 'koi_depth_err1',
 'koi_time0bk', 'koi_insol_err2', 'ra', 'koi_kepmag','koi_impact_err1','koi_steff','dec','koi_impact_err2','koi_srad','koi_slogg',
  'koi_srad_err2','koi_tce_plnt_num'], axis=1)
data2.head()
```

# Train the Model



```python
# Assign X (data) and y (target) on the second dataset
y = data2[["koi_disposition_CONFIRMED","koi_disposition_FALSE POSITIVE"]].values.reshape(-1, 2)
X = data2.drop(labels=["koi_disposition_CONFIRMED","koi_disposition_FALSE POSITIVE"], axis=1)
print(X.shape, y.shape)
```

```python
#Create a Train Test Split:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Scale your data
# Create a MinMaxScaler model and fit it to the training data
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
y_scaler = MinMaxScaler().fit(y_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
#Create a RandomForest model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train_scaled)
rf.score(X_test_scaled, y_test_scaled)
```

```python
print(f"Training Data Score: {rf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf.score(X_test_scaled, y_test)}")
```

```python
# Predicting the Test set results
y_pred = rf.predict(X_test_scaled)
```

```python
# Calculate classification report
target_names = ["confirmed", "false_positive"]
from sklearn.metrics import classification_report
print(classification_report(y_test_scaled, y_pred,
                            target_names=target_names))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
# Train the model with GridSearch
grid_search.fit(X_train_scaled, y_train_scaled)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```
